## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import random

#teórica:
cont_teorT=0
cont_teor1=0
cont_teor2=0

for i in range(1,7):
    for j in range(1,7):
        resultado=i+j
        cont_teorT +=1
        if resultado%2!=0: cont_teor1 +=1 
        if resultado>7: cont_teor2+=1

prob_teor1=cont_teor1/cont_teorT*100
#print(prob_teor1)
prob_teor2=round(cont_teor2/cont_teorT,2)*100
#print(prob_teor2)

def exercise1(x=1000):
    
    contex1=0
    contex2=0
    for i in range(x):
        dice1=random.randint(1,6)
        dice2=random.randint(1,6)
        sum=dice1+dice2
        if sum%2!=0: contex1 +=1
        if sum>7: contex2 +=1
    
    

    return print(f"Probabilidad del:\nExperimento 1: {round((contex1/x)*100,ndigits=2)}% ||Según teorema: {prob_teor1}%\nExperimento 2: {round(contex2/x,2)*100}% ||Según teorema: {prob_teor2}%"
               )

exercise1()

```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np
import random

ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)



#unique, counts=np.unique(x,return_counts=True)
#print(dict(zip(unique,counts)))

#Probabilidad según los teoremas de la probabilidad:
arr=list(np.array([ball_box[j] for j in range(len(ball_box))]))
prob_w=arr.count("White")/len(arr)
prob_r=arr.count("Red")/len(arr)
prob_g=arr.count("Green")/len(arr)

prob_exper1=round(prob_w**3*prob_r**2,4)*100
prob_exper2=round(prob_w**5+prob_g**5+prob_r**5,2)*100

print(prob_exper1,prob_exper2)

def exercise2(x=1000,n=5):
  
  
  counter1=0
  counter2=0

  for i in range(x):
     res=[]
  
     for i in range(n):
        var_rand=random.randint(0,len(ball_box)-1)
        res.append(ball_box[var_rand])
     w=res.count("White")
     r=res.count("Red")
     g=res.count("Green")

     if w==3 and r==2: counter1 += 1
     if w==5 or r==5 or g==5: counter2 += 1
     
  return print(f"Probabilidad del:\nExperimento 1: {round(counter1/x,3)*100}% ||Según teorema: {prob_exper1}%\nExperimento 2: {round(counter2/x,2)*100}% ||Según teorema: {prob_exper2}%"
               )

exercise2(x=100000)

print("\nAlgo estoy calculando mal de manera teórica ya que me sale 10 veces menos probable en el experimento 1")
```
